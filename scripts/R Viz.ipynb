```python
%%R
library(ggmap)
library(ggplot2)
library(tidyverse)
library(dplyr)
library(repr)
```

```python
%%R
mydf = read.csv('R_viz.csv')
map <- get_map(location = c(left = -100, bottom = 10,
                            right = -60, top = 40),
               source="stamen", maptype="terrain")
mydf = mydf %>% filter(Code == 'AL092017')
mydf1 = mydf %>% filter(Hour %% 6 == 0)
mydf2 = mydf %>% filter(Hour == 12)
```

```python
%%R -w 50 -h 40 -u cm
ggmap(map) +
geom_segment(data = mydf,
             aes(x = Lon1, y = Lat1, xend = Lon2, yend = Lat2, color=Bearing),  
             size = 6, 
             alpha = 0.8,
             lineend = "round")+
geom_point(data = mydf1, aes(x = Lon1, y = Lat1), alpha=0.2,size = 2,) +
geom_point(data = mydf2, aes(x = Lon1, y = Lat1), alpha=1,size = 2,) +
scale_colour_gradientn(colours = c("red","white","darkblue")) +
theme(axis.text=element_text(size=28),axis.title=element_text(size=34,face="bold"))
```
